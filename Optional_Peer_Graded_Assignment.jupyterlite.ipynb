## Questions

### Q1: Is there a significant difference in median value of houses bounded by the Charles river or not? 

```python
ax = sns.histplot(boston_df, x='MEDV', hue='CHAS', kde=True)
```

```python
bx = sns.boxplot(
    x=boston_df['CHAS'], 
    y=boston_df['MEDV']
)
```

**Hypothesis statement:**

*   $H\_0: µ\_1 = µ\_2$ ("there is no difference in median value between houses bounded by the Charles river and houses not bounded")
*   $H\_1: µ\_1 ≠ µ\_2$ ("there is a difference in median value between houses bounded by the Charles river and houses not bounded")

```python
scipy.stats.ttest_ind(
    boston_df[boston_df['CHAS']==1.0]['MEDV'],
    boston_df[boston_df['CHAS']==0.0]['MEDV']
)
```

**Conclusion:** Since the p-value is less than alpha value 0.05, we reject the null hypothesis as there is enough proof that there is a statistical difference in median value between houses bounded by the Charles river and houses not bounded

### Q2: Is there a difference in median values of houses of each proportion of owner-occupied units built before 1940?

```python
boston_df.loc[(boston_df['AGE'] <= 35.0), 'age_group'] = '35 years and younger'
boston_df.loc[(boston_df['AGE'] > 35.0)&(boston_df['AGE'] < 70.0), 'age_group'] = 'between 35 and 70 years'
boston_df.loc[(boston_df['AGE'] >= 70.0), 'age_group'] = '70 years and older'
```

```python
sns.boxplot(boston_df, y='MEDV', hue='age_group')
```

```python
scipy.stats.levene(boston_df[boston_df['age_group'] == '35 years and younger']['MEDV'],
                   boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV'], 
                   boston_df[boston_df['age_group'] == '70 years and older']['MEDV'], 
                   center='mean')
```

**Hypothesis statement:**

*   $H\_0: µ\_1 = µ\_2 = µ\_3$ (the three population means are equal)
*   $H\_1:$ At least one of the means differ

```python
scipy.stats.f_oneway(
    boston_df[boston_df['age_group'] == '35 years and younger']['MEDV'],
                   boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV'], 
                   boston_df[boston_df['age_group'] == '70 years and older']['MEDV']
)
```

**Conclusion:** Since the p-value is less than 0.05, we reject the null hypothesis as there is significant evidence that at least one of the means differ.

### Q3: Can we conclude that there is no relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town?

```python
ax = sns.scatterplot(x='NOX', y='INDUS', data=boston_df)
pyplot.show()
```

**Hypothesis Statement:**

*   $H\_0:$ Nitric oxide concentrations are not correlated with proportion of non-retail business acres per town
*   $H\_1:$ Nitric oxide concentrations are correlated with proportion of non-retail business acres per town


```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

**Conclusion:** Since p-value is smaller than 0.05, we reject the null hypothesis as there is significant evidence that there is a relationship between Nitric oxide consentrations and proportion of non-retail business acres per town
